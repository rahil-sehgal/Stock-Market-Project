```python
import yfinance as yf
import pandas as pd
import sqlalchemy
import datetime as dt
from pandas.tseries.offsets import MonthEnd
```

```python
wiki = 'https://en.wikipedia.org/wiki/'
tickersSensex = pd.read_html(wiki+'BSE_SENSEX')[1].Symbol.to_list()
```

```python
tickersSensex
```

```python
data = []
for ticker in tickersSensex:
    data.append(yf.download(ticker).reset_index())
```

```python
engine = sqlalchemy.create_engine('sqlite:///India.db')
```

```python
#data
```

```python
for frame, symbol in zip(data, tickersSensex):
    frame.to_sql(symbol, engine, index = False)
```

```python
df = pd.DataFrame()
```

```python
for name in tickersSensex:
    df = df.append(pd.read_sql(\
        f'SELECT Date, "Adj Close" AS "{name}" FROM "{name}" WHERE Date > "2010-01-10"',engine))
```

```python
df = df.groupby("Date").sum()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
# get this data frame and take the last price in a particular month
# month and price are enough
mtlprices = df.resample('M').last()
```

```python
mtlprices
```

```python
formation = dt.datetime(2011, 2, 28) 
# this is the date where i am forming my portfolio
```

```python
begin_measurement = formation - MonthEnd(12)
```

```python
end_measurement = formation - MonthEnd(1)
#only subtract one month due to short term reversal effect
```

```python
#check price at begin and end
price_end = mtlprices.loc[end_measurement]
```

```python
price_12mbefore = mtlprices.loc[begin_measurement]
```

```python
ret_12 = price_end/price_12mbefore - 1
#returns of past 12 months skipping the most recent month 
```

```python
ret_12
```

```python
winners = ret_12.nlargest(3)
```

```python
#now we will se if we hold portfolio for one month how our winners are performing 
winnerret = mtlprices.loc[formation+MonthEnd(1), winners.index] / mtlprices.loc[formation, winners.index] - 1

#one month after the formation period but only for winner stocks 
#we are calculating winner return so subtract one for that
```

```python
winnerret
```

```python
Momentumprofit = winnerret.mean()
```

```python
def momentumprofit(formation, holdingperiod = 1):
    begin_measurement = formation - MonthEnd(12)
    end_measurement = formation - MonthEnd(1)
    price_end = mtlprices.loc[end_measurement]
    price_12mbefore = mtlprices.loc[begin_measurement]
    ret_12 = price_end/price_12mbefore - 1
    winners = ret_12.nlargest(3)
    winnerret = mtlprices.loc[formation+MonthEnd(holdingperiod), winners.index] / mtlprices.loc[formation, winners.index] - 1
    #change the above MonthEnd(holdingperiod)
    Momentumprofit = winnerret.mean()
    return Momentumprofit
    
    
```

```python
profits, dates, holding = [], [], [] #3 empty list 
```

```python
#now populate them using nested loops
for i in (1,3,6): #i is the holding period
    for j in mtlprices.index[12:-i]: #we are always going 12 month back
        #populating the profits
        profits.append(momentumprofit(j, holdingperiod=i))
        dates.append(j + MonthEnd(i))
        holding.append(i)
        
```

```python
frame = pd.DataFrame({'Momentumprofit': profits, 'holdingperiod': holding}, index = dates)
```
